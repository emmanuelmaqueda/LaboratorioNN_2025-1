### Ejemplo: Relaci√≥n entre el n√∫mero de horas estudiadas y la calificaci√≥n obtenida en un examen.



```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Generar datos de ejemplo
np.random.seed(42)
horas_estudio = np.random.rand(100, 1) * 10  # Horas de estudio entre 0 y 10
calificaciones = 2.5 * horas_estudio + np.random.randn(100, 1) * 3 + 5  # Relaci√≥n lineal con ruido

# Visualizar los datos
plt.scatter(horas_estudio, calificaciones)
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificaciones")
plt.title("Datos de Ejemplo: Horas de Estudio vs Calificaciones")
plt.show()
```

```python
# Calcular las medias de X y Y
mean_x = np.mean(horas_estudio)
mean_y = np.mean(calificaciones)

# Calcular la pendiente (w1)
numerador = np.sum((horas_estudio - mean_x) * (calificaciones - mean_y))
denominador = np.sum((horas_estudio - mean_x) ** 2)
w1 = numerador / denominador

# Calcular el sesgo (b)
b = mean_y - w1 * mean_x

print(f"Pendiente (w1): {w1}")
print(f"Sesgo (b): {b}")
```

```python
# Hacer predicciones
calificaciones_pred = w1 * horas_estudio + b
mse_manual = np.mean((calificaciones - calificaciones_pred) ** 2)
print(f"Error Cuadr√°tico Medio (MSE): {mse_manual}")
```

```python
# Visualizar la l√≠nea de regresi√≥n
plt.scatter(horas_estudio, calificaciones, label="Datos Reales")
plt.plot(horas_estudio, calificaciones_pred, color='red', label="L√≠nea de Regresi√≥n Manual")
plt.xlabel("Horas de Estudio")
plt.ylabel("Calificaciones")
plt.title("Regresi√≥n Lineal Manual: Horas de Estudio vs Calificaciones")
plt.legend()
plt.show()
```

##Neurona Artificial y Neurona Biologica

Las redes neuronales artificiales, son algoritmos de software cuyo fundamento se inspira en el comportamiento de las neuronas del cerebro humano. En nuestro cerebro ocurren millones de reacciones qu√≠micas que comunican nuestras c√©lulas cerebrales, o neuronas, unas con otras, mediante estructuras especializadas.

image.png


Las neuronas se conectan entre s√≠ por medio de conexiones llamadas sinapsis, las cuales son intercambios el√©ctricos entre el bot√≥n sin√°ptico de una neurona (emisora o presin√°ptica) y el bot√≥n dendr√≠tico de otra (receptora o pos sin√°ptica). Dichas conexiones permiten que una neurona mande descargas el√©ctricas a las c√©lulas con las que se encuentra conectada.

La descarga el√©ctrica viaja desde el cuerpo o soma de una neurona hasta el Ax√≥n, el cual realiza la funci√≥n de un ca√±√≥n el√©ctrico a trav√©s de los neurotransmisores.
Cuando una neurona recibe un est√≠mulo su electroqu√≠mica se altera, acumulando energ√≠a. Cuando esa energ√≠a se acumula en cierta cantidad, la neurona la descarga por medio del Ax√≥n hacia las neuronas vecinas, formando una conexi√≥n sin√°ptica con ellas.
